```python
from ckiptagger import data_utils, construct_dictionary, WS, POS, NER
```

```python
ws=WS("./data")
pos=POS("./data")
ner=NER("./data")
```

```python
import pandas as pd
```

```python
df = pd.read_csv("台北市政府crawlerCKIP.csv")
```

```python
df = df.drop(["Unnamed: 0"],axis=1)
```

```python
df['data_description'][0]
```

```python
def Noun(target_string):
    result = []
    #get ws,pos,ner
    word_sentence_list = ws([target_string])
    pos_sentence_list = pos(word_sentence_list)
    entity_sentence_list = ner(word_sentence_list, pos_sentence_list)
    #select ORG
    for x in entity_sentence_list[0]:
        if x[2] == 'ORG':
             result.append(x[3])
    #select Noun
    for word_, pos_ in zip(word_sentence_list[0], pos_sentence_list[0]):
        if (pos_ == 'Na') or (pos_ == 'Nb') or (pos_ == 'Nc'):
            result.append(word_)
    result = ','.join(list(set(result)))
    return result
```

```python
df['key_words'] = df['data_description'].apply(lambda x : Noun(x))
```
